```python
class Book:
    def __init__(self, title, author, price, quantity, book_type):
        self.title = title #book title
        self.author = author #book author
        self.price = price #book price
        self.quantity = quantity #book quanity
        self.type = book_type #book type genere
        self.observers = set() #Setting price and all book in one variable into dictionary

    def addObserver(self, observer):
        self.observers.add(observer)

    def removeObserver(self, observer):
        self.observers.remove(observer)

    def notifyObservers(self):
        for observer in self.observers:
            observer.update(self)

    def getTitle(self):
        return self.title

    def getAuthor(self):
        return self.author

    def getPrice(self):
        return self.price

    def getQuantity(self):
        return self.quantity

    def getType(self):
        return self.type

    def setPrice(self, newPrice):
        self.price = newPrice
        self.notifyObservers()

    def setQuantity(self, newQuantity):
        self.quantity = newQuantity
        self.notifyObservers()

    def __repr__(self):
        return f"Book({self.title}, {self.author}, {self.price}, {self.quantity}, {self.type})"


class User:
    def __init__(self, username, email, password):
        self.username = username #USERNAME  REGISTRATION VIA THIS WE WILL CHECK THE SIGN IN PROCESS
        self.email = email  #FOR SIGN IN PROCESS WE WILL USE ONLY USER ENTERED EMAIL AND PASSWORD VIA BOTH THESE USER CAN SIGN IN
        self.password = password

    def getUsername(self):
        return self.username

    def getEmail(self):
        return self.email

    def setEmail(self, newEmail):
        self.email = newEmail

    def setPassword(self, newPassword):
        self.password = newPassword

    def __repr__(self):
        return f"User({self.username}, {self.email})"


class Order:
    def __init__(self, order_id, user, book, quantity):
        self.order_id = order_id
        self.user = user
        self.book = book
        self.quantity = quantity

    def getOrderId(self):
        return self.order_id  #GETORDER ID WHEN EVER USER PURCHASED

    def getBook(self):
        return self.book   #BOOKS NAME

    def getQuantity(self):
        return self.quantity  #BOOKS QUANTITY

    def setQuantity(self, newQuantity):
        self.quantity = newQuantity  #AFTER USER PURCHASE WE ARE SETTING NEW QUANTITY FOR BOOK

    def __repr__(self):
        return f"Order({self.order_id}, {self.user}, {self.book}, {self.quantity})"


class BookObserver:
    def update(self, book):
        print("-------------------------------------------------------------------------------------------------------------------------------------------------------------")
        print(f"New Book Added: {book.getTitle()} by {book.getAuthor()}") #PRINTING FIRST NEW BOOK HERE TO SHOW CASE THE USER
        print("-------------------------------------------------------------------------------------------------------------------------------------------------------------")


class OnlineBookstore:
    def __init__(self):
        self.books = []
        self.users = []
        self.orders = []

    def display_books(self):
        print("Books Available")
        for i, book in enumerate(self.books, start=1):
            print(f"{i}. {book.getTitle()} by {book.getAuthor()} - ${book.getPrice()}")

    def order_book(self, user, book, quantity):   #order a book here function
        if book.getQuantity() >= quantity:  #book quantity
            order_id = str(len(self.orders) + 1)
            order = Order(order_id, user, book, quantity)
            self.orders.append(order)
            book.setQuantity(book.getQuantity() - quantity)
            print(f"Successfully Ordered Your Books - Order ID: {order_id}")
        else:
            print("Stock Insufficient. Choose a lower quantity.")   #IF BOOK QUANTITY  IS NOT ENOUGH

    def register_user(self, username, email, password):   #REGISTERING USER HERE
        user = User(username, email, password)
        self.users.append(user)
        print("Successfully Registered!")

    def sign_in(self, email, password):  #SIGN IN FUNCTION WHERE WE ARE CHECKING USER CREDENTIAL
        for user in self.users:
            if user.getEmail() == email and user.password == password:
                print(f"Welcome, {user.getUsername()}!")
                return user
        print("Invalid Credentials. Try again or Register Yourself")

    def add_book(self, book):   #ADDING NEW BOOKS HERE
        self.books.append(book)
        for observer in book.observers:
            observer.update(book)


# Example usage:
if __name__ == "__main__":
    online_bookstore = OnlineBookstore()   #ALL THESE BOOKS IS RANDOM WE CAN CHANGE ACCORDING TO REQUIREMENTS
    book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", 15.99, 100, "Fiction") #BOOKNAME, AUTHOR NAME, BOOK PRICE, BOOK QUANTITY AVAILBLE, BOOK TYPE
    book2 = Book("The Hobbit", "J.R.R. Tolkien", 19.99, 50, "Fantasy")  #ALL BOOKS ARE ARRANGED ACCORDING TO ABOVE ORDER
    book3 = Book("To Kill a Mockingbird", "Harper Lee", 12.99, 80, "Fiction")
    book4 = Book("1984", "George Orwell", 14.99, 120, "Dystopian")
    book5 = Book("The Catcher in the Rye", "J.D. Salinger", 11.99, 90, "Coming-of-age")
    book6 = Book("The Lord of the Rings", "J.R.R. Tolkien", 24.99, 60, "Fantasy")
    book7 = Book("The Da Vinci Code", "Dan Brown", 17.99, 110, "Mystery")
    book8 = Book("Pride and Prejudice", "Jane Austen", 13.99, 70, "Romance")
    book9 = Book("Harry Potter and the Sorcerer's Stone", "J.K. Rowling", 21.99, 40, "Fantasy")
    book10 = Book("The Shining", "Stephen King", 16.99, 85, "Horror")
    book11 = Book("The Chronicles of Narnia", "C.S. Lewis", 18.99, 75, "Fantasy")
    book12 = Book("The Hitchhiker's Guide to the Galaxy", "Douglas Adams", 16.99, 95, "Sci-Fi")
    book13 = Book("The Girl on the Train", "Paula Hawkins", 14.99, 110, "Mystery")
    book14 = Book("The Alchemist", "Paulo Coelho", 20.99, 60, "Philosophical Fiction")
    book15 = Book("The Martian", "Andy Weir", 22.99, 45, "Sci-Fi")
    book16 = Book("The Art of War", "Sun Tzu", 11.99, 120, "Philosophy")
    book17 = Book("The Silent Patient", "Alex Michaelides", 19.99, 80, "Thriller")
    book18 = Book("The Hunger Games", "Suzanne Collins", 14.99, 100, "Dystopian")
    book19 = Book("The Fault in Our Stars", "John Green", 12.99, 90, "Romance")
    book20 = Book("The Road", "Cormac McCarthy", 15.99, 70, "Post-apocalyptic Fiction")
    #AFTER THAT WE WILL STORE ALL BOOKS IN BOOKS VARIABLE
    books = [book1, book2, book3, book4, book5, book6, book7, book8, book9, book10,book11, book12, book13, book14, book15, book16, book17, book18, book19, book20]


    observer = BookObserver()
    for book in books:
      book.addObserver(observer)
      online_bookstore.add_book(book)
#BOOK CLASS  THIS IS THE BLUEPRINT OF BOOKS.
#IT IS STORING ALL THE INFORMATION ABOUT A BOOK

#MENU DRIVEN APPROACH START
#WE HAVE TO CHOOSE ONE OPTION FROM HERE
    while True:
        print("---------------------------------ONLINE--BOOK--STORE-----------------------------------------")
        print("\n1. Display Books\n2. Order Book\n3. Register\n4. Sign In\n5. Exit")
        choice = input("Choose one option from the above list: ")

#1 DISPLAY ALL THE BOOKS AVAILBLE IN THE STORE
        if choice == "1":
            online_bookstore.display_books()
        elif choice == "2":   #FOR ORDERING THE BOOKS
            if not online_bookstore.users:
                print("Please sign in or register first before purchasing")   #YOU HAVE TO SIGN IN OR REGISTER FIRST , IF YOU ARE ORDERING BOOKS FROM BOOKS STORE
                continue

            online_bookstore.display_books()
            book_choice = int(input("Enter the book number to order for purchasing:- ")) - 1 #USER INPUT - 1
            quantity = int(input("Enter the quantity for purchasing: ")) #QAUNTITY OF BOOKS , NOT GREATER THAN BOOK STORE
            online_bookstore.order_book(online_bookstore.users[0], online_bookstore.books[book_choice], quantity)
        elif choice == "3": # FOR 3 WE HAVE REGISTRATION PROCESS
            print("REGISTRATION")  #FOR THAT USER HAVE TO INSERT THEIR EMAIL USERNAME AND PASSWORD
            username = input("Enter your username: ")
            email = input("Enter your email: ")
            password = input("Enter your password: ")
            online_bookstore.register_user(username, email, password)
        elif choice == "4":   #SIGN IN MODEL
            print("SIGN-IN")  #FOR SIGN IN WE ARE USING USER EMAIL AND PASSWORD ONLY
            email = input("Enter your email: ")
            password = input("Enter your password: ")
            signed_in_user = online_bookstore.sign_in(email, password)
            if signed_in_user:
                print(f"Signed in as {signed_in_user.getUsername()}")
        elif choice == "5":   #WRONG CHOICE
            break
        else:
            print("Invalid choice. Please try again and choose a valid number")
```
